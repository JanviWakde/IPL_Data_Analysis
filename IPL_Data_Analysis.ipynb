```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches_df = pd.read_csv(r"C:\Users\wakde\DemoPython\matches.csv")
deliveries_df = pd.read_csv(r"C:\Users\wakde\DemoPython\deliveries.csv")
```

```python
matches_df.head(5)
```

```python
# Removing the Umpire3 column as all its values are NaN
matches_df.drop(columns=['umpire3'], inplace = True)
```

```python
total_num_of_matches = len(matches_df.index)
print(total_num_of_matches)
```

```python
# Top 10 batsmen with runs
b_runs = deliveries_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
top_10_batsmen = b_runs.head(10)
print(top_10_batsmen)
```

```python
# Wickets by bowler
wickets= deliveries_df[deliveries_df['player_dismissed'].notna()]
b_wkts = wickets.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False)
top_10_bowlers = b_wkts.head(10)
print(top_10_bowlers)
```

```python
# Matches Won per Team per Year
wins = matches_df.groupby(['season','winner'])['id'].count().reset_index(name='wins')
pivot = wins.pivot(index = 'season', columns = 'winner', values = 'wins').fillna(0)
print(pivot)
```

```python
# Total runs per season for top batsmen
runs_season = deliveries_df.merge(matches_df[['id', 'season']], left_on='match_id', right_on='id')
runs_by_season = runs_season.groupby(['season', 'batsman'])['batsman_runs'].sum().reset_index()
top_per_season = runs_by_season.groupby('season').apply(lambda x: x.nlargest(3, 'batsman_runs')).reset_index(drop=True)
print(top_per_season)
```

```python
# Top 10 Batsmen
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_batsmen.values, y=top_10_batsmen.index, palette='viridis')
plt.title('Top 10 Run Scorers in IPL')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.show()
```

```python
# Top 10 Bowlers
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_bowlers.values, y=top_10_bowlers.index, palette='magma')
plt.title('Top 10 Wicket Takers in IPL')
plt.xlabel('Total Wickets')
plt.ylabel('Bowler')
plt.show()
```

```python
# Team Wins at Each Venue
venue_wins = matches_df.groupby(['venue', 'winner']).size().unstack().fillna(0)

plt.figure(figsize=(14, 10))
sns.heatmap(venue_wins, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title('Team Wins at Each Venue')
plt.xlabel('Team')
plt.ylabel('Venue')
plt.show()
```

```python
# Number of matches played in each season
plt.figure(figsize=(13,8))
plot = sns.countplot(x = 'season', data = matches_df, palette=sns.color_palette('cubehelix',10))
plt.ylabel('Number of matches')
plt.title('Number of matches played in each season')
plt.show()
```

```python
#Total runs in each season
merged_df = matches_df[['id','season']].merge(deliveries_df, left_on = 'id', right_on = 'match_id', how = 'left')
season_runs = merged_df.groupby(['season'])['total_runs'].sum().reset_index()
plt.subplots(figsize=(13,8))
plot = sns.barplot(x='season',y='total_runs',data=season_runs)
plot.set_xlabel('Season')
plot.set_ylabel('Total Runs')
plot.set_title('Total runs scored in each season')
```

```python
#Total matches won by each team
plt.subplots(figsize=(13,8))
sns.countplot(x='winner',data=matches_df)
plt.ylabel('Number Of Maches Won')
plt.xlabel('')
plt.xticks(rotation = 90)
```

```python
plt.subplots(figsize=(13,8))
sns.countplot(x='season',hue='toss_decision',data=matches_df,palette=sns.color_palette('Paired'))
```

```python
print("Winners by years")
winners = matches_df.drop_duplicates(subset=['season'], keep='last')[['season', 'winner']]
winners = winners.sort_values(by='season').reset_index(drop=True)
winners
```

```python
#compare two teams
def team_comparator(team1,team2):
    compare = matches_df[((matches_df['team1']==team1) | (matches_df['team2']==team1)) & ((matches_df['team1']==team2) | (matches_df['team2']==team2))]
    plt.subplots(figsize=(13,8))
    plt.yticks(np.arange(0,4,1))
    plot = sns.countplot(x='season', hue='winner',data = compare,palette=sns.color_palette('Set2'))
    plot.set_xlabel('Season')
    plot.set_ylabel('Win count')
team_comparator('Mumbai Indians','Chennai Super Kings')
```

```python
#Types of dismissal
plt.figure(figsize=(13,8))
plot = sns.countplot(x='dismissal_kind', data=deliveries_df)
plot.set_xlabel('')
plot.set_ylabel('Count')
plt.title('Kind of dismissal')
plt.yticks(np.arange(0,5000,500))
plt.show()
```

```python
#Bowlers with most number of wickets
top_bowlers = deliveries_df[deliveries_df["dismissal_kind"].isin(["bowled", "caught", "lbw", "stumped", "caught and bowled"])]
plt.subplots(figsize=(13,8))
plot = sns.countplot(y='bowler',data=top_bowlers,order=top_bowlers['bowler'].value_counts().iloc[:10].index,palette='RdPu')
plot.set_title('Bowlers with most Wickets')
```
